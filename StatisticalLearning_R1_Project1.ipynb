```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
laliga = pd.read_csv("Laliga.csv", header=1, index_col='Pos')  
laliga.replace('-', 0, inplace=True)
laliga.tail()
```

```python
debut = laliga.Debut.str.split('-', n = 1, expand = True) 
team3080 = laliga[(debut[0].astype(int) >= 1930) & (debut[0].astype(int) <= 1980)]
team3080['Team'].head()
```

```python
laliga['Points'] = pd.to_numeric(laliga['Points'])
laliga.sort_values(by='Points', ascending=False).head()['Team'].tolist()
```

```python
def Goal_diff_count(goalsFor, goalsAgainst):
    return (goalsFor - goalsAgainst)

laliga['GoalsFor'] = pd.to_numeric(laliga['GoalsFor'])
laliga['GoalsAgainst'] = pd.to_numeric(laliga['GoalsAgainst'])

laliga["GoalDiff"] = Goal_diff_count(laliga['GoalsFor'], laliga['GoalsAgainst'])
laliga[['Team', 'GoalDiff']].head()
```

```python
print('Max Goal Difference:', laliga.loc[laliga['GoalDiff'].idxmax(), 'Team'])
print('Min Goal Difference:', laliga.loc[laliga['GoalDiff'].idxmin(), 'Team'])
```

```python
laliga.drop(["GoalDiff"], axis = 1, inplace=True)
```

```python
def percentageOfWinning(GamesWon, GamesPlayed):
    try:
        return ((GamesWon / GamesPlayed)*100)
    except ArithmeticError:
        return 0

laliga['GamesWon'] = pd.to_numeric(laliga['GamesWon'])
laliga['GamesPlayed'] = pd.to_numeric(laliga['GamesPlayed'])

laliga['Winning Percent'] = percentageOfWinning(laliga['GamesWon'], laliga['GamesPlayed'])
laliga.head()
```

```python
laliga.sort_values(by='Winning Percent', ascending=False).head()['Team']
```
